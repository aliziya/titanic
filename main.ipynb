```python
import pandas as pd
```

```python
train_Data = pd.read_csv("Data/train.csv")
test_Data = pd.read_csv("Data/test.csv")

gender_submission = pd.read_csv("Data/gender_submission.csv")

test_Data = pd.merge(test_Data , gender_submission)
test_Data
```

```python
train_Data.isna().sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
non_numeric_feutures = ["Sex" , "Cabin" , "Embarked" , "Name" , "Ticket"]

imp =  SimpleImputer(strategy = "constant" , fill_value = "missing")
onehot = OneHotEncoder(handle_unknown = "ignore")

non_numeric_transformer = Pipeline(steps = [
    
    ("imputer" , SimpleImputer (strategy = "constant" , fill_value = "missing")) ,
    ("onehot" , OneHotEncoder (handle_unknown = "ignore"))
 ])
```

```python
preprocessor = ColumnTransformer ( [("non_numeric" , non_numeric_transformer , non_numeric_feutures)] )
```

```python
from sklearn.svm import LinearSVC

X_train = train_Data.drop(axis = 1 , columns= ["Survived"])
Y_train = train_Data ["Survived"]

X_test = test_Data.drop(axis = 1 , columns= ["Survived"])
Y_test = test_Data["Survived"]

model = LinearSVC()

model_pipeline = Pipeline (steps = [
    ( "preprocessor" , preprocessor ) ,
    ( "model" , model)
] )

model_pipeline.fit (X_train , Y_train)
model_pipeline.score (X_test , Y_test)
```
